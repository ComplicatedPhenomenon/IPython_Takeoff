```python
from __future__ import print_function

# Import conventions we'll be using here. See Part 1
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
qq=['../result/qqd2.txt','../result/qqd3.txt','../result/qqd4.txt','../result/qqd5.txt','../result/qqd6.txt']
gg=['../result/ggd2.txt','../result/ggd3.txt','../result/ggd4.txt','../result/ggd5.txt','../result/ggd6.txt']
data = [qq,gg]
mark = ['>', '*', 'x', 'd', '+']
tag = ['$\delta$ = 2','$\delta$ = 3','$\delta$ = 4','$\delta$ = 5','$\delta$ = 6']
font = {'family': 'serif',
        'color':  'darkred',
        'weight': 'normal',
        'size': 16,
        }
for j in range(0,2):
    for i in range(0,5):
        datafile = open(data[j][i],'r')
        dataz =  datafile.readlines()
        datafile.close()

        s = []
        cs = []
        for row in dataz:
            this_data = row.split()
            s.append(float(this_data[0]))
            cs.append(float(this_data[1]))


        S = np.array(s)
        CS = np.array(cs)

        plt.plot(S,1/4*CS*10**3,mark[i], label = tag[i])
        plt.xlabel("$S$ [GeV]",fontdict = font)
        plt.ylabel("$\sigma$ [fb]",fontdict = font)
        plt.legend()

    plt.yscale('log')
    plt.show()
```

```python
qq=['../result/qqd2.txt','../result/qqd3.txt','../result/qqd4.txt','../result/qqd5.txt','../result/qqd6.txt']
gg=['../result/ggd2.txt','../result/ggd3.txt','../result/ggd4.txt','../result/ggd5.txt','../result/ggd6.txt']
mark = ['>', '*', 'x', 'd', '+']
tag = ['$\delta$ = 2','$\delta$ = 3','$\delta$ = 4','$\delta$ = 5','$\delta$ = 6']
font = {'family': 'serif',
        'color':  'darkred',
        'weight': 'normal',
        'size': 16,
        }

    
for i in range(0,5):
    dataqq = open(qq[i],'r')
    datagg = open(gg[i],'r')
    data1 =  dataqq.readlines()
    data2 =  datagg.readlines()
    dataqq.close()
    datagg.close()

    s1 = []
    cs1 = []
    for row in data1:
        this_data = row.split()
        s1.append(float(this_data[0]))
        cs1.append(float(this_data[1]))
        
        
    s2 = []
    cs2 = []
    for row in data2:
        this_data = row.split()
        s2.append(float(this_data[0]))
        cs2.append(float(this_data[1]))
        
    S = np.array(s1)
    CS = np.array(cs1)+np.array(cs2)

    plt.plot(S,1/4*CS*10**3,mark[i], label = tag[i])
    plt.xlabel("$S$ [GeV]",fontdict = font)
    plt.ylabel("$\sigma$ [fb]",fontdict = font)
    plt.legend()

plt.yscale('log')
plt.show()
```
